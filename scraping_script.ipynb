# Convertion to txt

```python
import os
from docx import Document

# Specify the directory containing the Word documents
docx_directory = 'path/to/word/documents'

# Specify the output file path
output_file = 'path/to/save/output.txt'

# Open the output file in append mode
with open(output_file, 'a', encoding='utf-8') as txt_file:
    # Loop through each file in the docx directory
    for filename in os.listdir(docx_directory):
        if filename.endswith('.docx'):
            # Construct the full path of the input file
            docx_path = os.path.join(docx_directory, filename)

            # Open the Word document
            doc = Document(docx_path)

            # Extract the text from the document
            text = ' '.join([paragraph.text for paragraph in doc.paragraphs])

            # Write the text to the output file
            txt_file.write(text + '\n')

            print(f'Processed {filename}.')

print('Conversion complete.')
```
